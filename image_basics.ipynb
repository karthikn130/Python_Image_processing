```python
import numpy as np
import matplotlib.pyplot as plt 


from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python
np.arange(0,255,5)
array_spectrum = np.array([np.arange(0,255,5), np.arange(255,0,-5), np.arange(0,255,5), np.arange(255,0,-5)])
array_spectrum
```

```python
array_spectrum = np.array([np.arange(0,255,5), np.arange(255,0,-5), np.arange(0,255,5), np.arange(255,0,-5)])
fig, ax = plt.subplots(1, 2,figsize=(12,4))
ax[0].imshow(array_spectrum, cmap = 'gray')
ax[0].set_title('Arange Generation')
ax[1].imshow(array_spectrum.T, cmap = 'gray')
ax[1].set_title('Transpose Generation')
```

```python
doggo = imread('doggo.png')
imshow(doggo)
```

```python
from PIL import Image
doggy = Image.open("doggo.png")
# type(doggy)
print(doggy.mode)
```

```python
print("shape of matrix", doggo.shape)
print("dimensions", doggo.ndim)
print(doggo[0,0])
doggo[:,:,3] = 255
imshow(doggo)
```

```python
imshow(doggo)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(6,4),
sharey= True)
ax[0].imshow(doggo[:, 0:300])
ax[0].set_title('First Split')
ax[1].imshow(doggo[:, 300:600])
ax[1].set_title('Second Split')
ax[2].imshow(doggo[:, 600:1000])
ax[2].set_title('Third Split');
```

```python
imshow(doggo[0:600, 800:1400])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(doggo[:,:,0], cmap='Reds')
ax[0].set_title('Red')
ax[1].imshow(doggo[:,:,1],cmap='Greens')
ax[1].set_title('Green')
ax[2].imshow(doggo[:,:,2], cmap='Blues')
ax[2].set_title('Blue');
```

```python
doggo_hsv = rgb2hsv(doggo)
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(doggo_hsv[:,:,0], cmap='hsv')
ax[0].set_title('Hue')
ax[1].imshow(doggo_hsv[:,:,1], cmap='gray')
ax[1].set_title('Saturation')
ax[2].imshow(doggo_hsv [:,:,2], cmap='gray')
ax[2].set_title('Value');
```
